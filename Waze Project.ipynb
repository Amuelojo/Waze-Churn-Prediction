```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r'C:\Users\user\OneDrive\Desktop\Google Data Analytic Certification File\Waze Project\Waze_dataset.csv')

# Remove rows with missing values in the 'label' column
df_cleaned_specific = df.dropna(subset=['label'])

# Create a scatterplot of 'drives' vs 'sessions'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='drives', y='sessions', hue='label', data=df_cleaned_specific)
plt.title('Scatterplot of Drives vs Sessions')
plt.xlabel('Drives')
plt.ylabel('Sessions')
plt.legend(title='User Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\user\OneDrive\Desktop\Google Data Analytic Certification File\Waze Project\Waze_dataset.csv')

# Check the first few rows of the dataset to ensure it is loaded correctly
print(df.head())

# Set up the figure size
plt.figure(figsize=(15, 10))

# Distribution of Sessions
plt.subplot(2, 2, 1)
sns.histplot(df['sessions'], bins=50, kde=True)
plt.title('Distribution of Sessions')
plt.xlabel('Sessions')
plt.ylabel('Frequency')

```

```python
# Boxplot for Duration of Drives
plt.subplot(2, 2, 3)
sns.boxplot(y=df['duration_minutes_drives'])
plt.title('Boxplot of Drive Duration (in minutes)')
plt.ylabel('Drive Duration (minutes)')
```

```python
# Distribution of Drives
plt.subplot(2, 2, 2)
sns.histplot(df['drives'], bins=50, kde=True)
plt.title('Distribution of Drives')
plt.xlabel('Drives')
plt.ylabel('Frequency')
```

```python
# Scatter plot for Drives vs. Driven KM
plt.subplot(2, 2, 4)
sns.scatterplot(x='drives', y='driven_km_drives', data=df)
plt.title('Scatter plot of Drives vs. Driven KM')
plt.xlabel('Number of Drives')
plt.ylabel('Driven KM')
```

```python
# Show the plots
plt.tight_layout()
plt.show()
```

```python

# Bar plot for device type distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='device', data=df)
plt.title('Distribution of Users by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Count')
plt.show()
```
